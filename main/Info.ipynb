```python
import matplotlib.pyplot as plt
```

```python
sizes = []
times = []

with open("consumed_time.txt", "r") as file:
    for line in file:
        parts = line.split()
        sizes.append(int(parts[2]))
        times.append(float(parts[5]))

plt.figure(figsize=(10, 5))
plt.plot(sizes, times, marker='o')
plt.xlabel('Size')
plt.ylabel('Time, sec')
plt.title('Time vs Size')
plt.grid()
plt.show()
```
